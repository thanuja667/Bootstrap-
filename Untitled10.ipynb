```r
library(ggplot2)
library(dplyr)
library(tidyr)
```

```r
df <- read.csv("mental.csv", stringsAsFactors = FALSE)
head(df)
```

```r
list.files()
```

```r
#mental helath analysis of tech employees
install.packages(c("ggplot2", "dplyr", "caret", "tidyr"), dependencies = TRUE)
```

```r
# Filter valid age range
df <- df[df$Age >= 18 & df$Age <= 65, ]

# Clean Gender column
df$Gender <- tolower(df$Gender)
df$Gender <- ifelse(grepl("female", df$Gender), "Female",
             ifelse(grepl("male", df$Gender), "Male", "Other"))

# Convert relevant columns to factor
#These categorical variables are converted into factor type, which is essential for modeling and plotting.
df$treatment <- as.factor(df$treatment)
df$Gender <- as.factor(df$Gender)
df$remote_work <- as.factor(df$remote_work)
df$benefits <- as.factor(df$benefits)
df$seek_help <- as.factor(df$seek_help)
df$wellness_program <- as.factor(df$wellness_program)
df$care_options <- as.factor(df$care_options)
df$anonymity <- as.factor(df$anonymity)
df$family_history <- as.factor(df$family_history)
```

```r
head(df)
```

```r
# Step 4: Convert categorical columns to factors data cleaning and visualization
factor_cols <- c("treatment", "Gender", "remote_work", "benefits",
                 "seek_help", "wellness_program", "care_options",
                 "anonymity", "family_history")
for (col in factor_cols) {
  df[[col]] <- as.factor(df[[col]])
}
cat("✅ Step 4: Columns converted to factors\n")
```

```r
# Histogram: Age distribution
ggplot(df, aes(x = Age)) +
  geom_histogram(binwidth = 2, fill = "steelblue", color = "black") +
  labs(title = "Age Distribution of Tech Employees", x = "Age", y = "Count") +
  theme_minimal()
```

```r
# Boxplot: Age vs Treatment
ggplot(df, aes(x = treatment, y = Age, fill = treatment)) +
  geom_boxplot() +
  labs(title = "Age vs Mental Health Treatment") +
  theme_minimal()
```

```r
# Stacked Bar Plot: Treatment by Gender
ggplot(df, aes(x = Gender, fill = treatment)) +
  geom_bar(position = "stack") +
  labs(title = "Mental Health Treatment by Gender") +
  theme_minimal()
```

```r
# Heatmap: Work Interference vs Treatment
work_treatment_matrix <- table(df$work_interfere, df$treatment)

# Convert the table to a data frame
work_treatment_df <- as.data.frame(as.table(work_treatment_matrix))

ggplot(work_treatment_df, aes(x = Var2, y = Var1, fill = Freq)) +
  geom_tile() +
  scale_fill_gradient(low = "lightblue", high = "darkblue") +
  labs(title = "Heatmap: Work Interference vs Mental Health Treatment") +
  theme_minimal()
```

```r
# Bar Plot: Benefits vs Treatment
ggplot(df, aes(x = benefits, fill = treatment)) +
  geom_bar(position = "dodge") +
  labs(title = "Mental Health Benefits vs Treatment") +
  theme_minimal()
```

```r
# Density Plot: Age by Treatment
ggplot(df, aes(x = Age, fill = treatment)) +
  geom_density(alpha = 0.5) +
  labs(title = "Age Density by Treatment") +
  theme_minimal()
```

```r
install.packages("caret")  # Only once if not already installed
library(caret)             # Every time before using createDataPartition
```

```r
# Load required library
library(caret)

# Select relevant features
model_data <- df %>%
  select(Age, Gender, remote_work, benefits, seek_help, family_history, treatment) %>%
  drop_na()

# Ensure all are factors
#Ensures all predictors are factors as required by the logistic regression model.
model_data$Gender <- as.factor(model_data$Gender)
model_data$remote_work <- as.factor(model_data$remote_work)
model_data$benefits <- as.factor(model_data$benefits)
model_data$seek_help <- as.factor(model_data$seek_help)
model_data$family_history <- as.factor(model_data$family_history)

# Split the dataset
set.seed(123)
splitIndex <- createDataPartition(model_data$treatment, p = 0.7, list = FALSE)
train <- model_data[splitIndex, ]
test <- model_data[-splitIndex, ]

# Train model
log_model <- glm(treatment ~ ., data = train, family = "binomial")
summary(log_model)

# Predict and evaluate
preds <- predict(log_model, test, type = "response")
pred_labels <- ifelse(preds > 0.5, "Yes", "No")
confusionMatrix(as.factor(pred_labels), test$treatment)
```

```r
# Step 7: Train model
log_model <- tryCatch({
  glm(treatment ~ ., data = train, family = "binomial")
}, error = function(e) {
  cat("❌ Step 7 Failed: Model training error\n")
  return(NULL)
})

if (!is.null(log_model)) {
  cat("✅ Step 7: Logistic regression model trained\n")
}
```

```r
# Step 8: Predict and evaluate
preds <- predict(log_model, test, type = "response")
pred_labels <- ifelse(preds > 0.5, "Yes", "No")
confusionMatrix(as.factor(pred_labels), test$treatment)
cat("✅ Step 8: Predictions and evaluation complete\n")
```
